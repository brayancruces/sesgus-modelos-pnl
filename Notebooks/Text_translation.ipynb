```python
from googletrans import Translator
translator = Translator()
```

```python
result=translator.translate('Men cannot relate to other persons as well as', dest='es')
print(result.text)
```

```python
def get_spanish(value):
  result=translator.translate(value, dest='es')
  return result.text

df['spanish_text'] = df['text'].map(get_spanish)
```

```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()


similar('colegiala','Chica de escuela')
```

```python

arr_index_similar=[]
final_arr=[]
for i in range(len(df_ner['spanish_text'])):
  for sub in list(df_ner['spanish_text'])[i]:
    print(similar(sub, list(df_ner['target_es'])[i]))

    arr_index_similar.append(sub)
  final_arr.append(arr_index_similar)
  
    
```
