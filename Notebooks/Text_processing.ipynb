```python
# Cargando datos
biased_data='/content/drive/MyDrive/Hackathones/BBVA Hakathon 2022/Bases de datos/CMS Dataset/Subset/data_convertida_esp.csv'
not_baised='/content/drive/MyDrive/Hackathones/BBVA Hakathon 2022/Bases de datos/CMS Dataset/Subset/data_convertida_notbiased_esp.csv'
bias=pd.read_csv(biased_data)
notbias=pd.read_csv(not_baised)


# Cargando datos
nontoxic='/content/drive/MyDrive/Hackathones/BBVA Hakathon 2022/Bases de datos/CMS Dataset/Subset/nontoxic_data_convertida_notbiased_esp.csv'
toxic='/content/drive/MyDrive/Hackathones/BBVA Hakathon 2022/Bases de datos/CMS Dataset/Subset/toxic_data_convertida_esp.csv'
nontoxic=pd.read_csv(nontoxic)
toxic=pd.read_csv(toxic)

toxic['label']='negative'
nontoxic['label']='positive'
```

```python
def remove_urls(row): 
    text=row['spanish_text']
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    a=url_pattern.sub(r'', text)
    return a
```

```python
def demojir(row):
    text=row['spanish_text']
    text=demoji.replace(text)
    return text
```

```python
def remove_user(row):
    text=row['spanish_text']
    r = re.findall("@[\w]*", text)
    for i in r:
        text = re.sub(i, '', text)
    return text
```

```python
def remove_hastag(row):
    text=row['spanish_text']
    r = re.findall("#[\w]*", text)
    for i in r:
        text = re.sub(i, '', text)
    return text
    
```

```python
def remove_rt(row):
    text=row['spanish_text']
    r = re.findall('RT [\w]*', text)
    for i in r:
        text = re.sub(i, '', text)
    return text.lower()

def remove_mt(row):
    text=row['spanish_text']
    r = re.findall('mention[\w]*', text)
    #print(r)
    for i in r:
      text = re.sub(i, '', text)
    return text.lower()

def remove_mc(row):
    text=row['spanish_text']
    r = re.findall('menc[\w]*', text)
    #print(r)
    for i in r:
      text = re.sub(i, '', text)
    return text.lower()

def remove_number(row):
    text=row['spanish_text']
    r = re.findall('[\d]+', text)
    #print(r)
    for i in r:
      text = re.sub(i, '', text)
    return text

    
```

```python
def signos(row):

    sentences=row['spanish_text']
    text_string=re.findall('[\w]+',sentences)
    cad=' '.join(text_string)
    cad=unidecode.unidecode(cad)
    cad2=cad.translate(str.maketrans("",'',a))
    cad3=cad2.split()
    cad4=' '.join(cad3)
    
    return cad4
```

```python
bias=bias[['spanish_text','sexist']]
notbias=notbias[['spanish_text','sexist']]



nontoxic=nontoxic [['spanish_text','label']]
toxic=toxic[['spanish_text','label']]
```

```python
bias['spanish_text'] = bias.apply(demojir, axis=1)
bias['spanish_text'] = bias.apply(remove_user, axis=1)
bias['spanish_text'] = bias.apply(remove_hastag, axis=1)
bias['spanish_text'] = bias.apply(remove_rt, axis=1)
bias['spanish_text'] = bias.apply(remove_mt, axis=1)
bias['spanish_text'] = bias.apply(remove_mc, axis=1)
bias['spanish_text'] = bias.apply(remove_number, axis=1) 
bias['spanish_text'] = bias.apply(remove_urls, axis=1)
bias['spanish_text'] = bias.apply(signos, axis=1)


notbias['spanish_text'] = notbias.apply(demojir, axis=1)
notbias['spanish_text'] = notbias.apply(remove_user, axis=1)
notbias['spanish_text'] = notbias.apply(remove_hastag, axis=1)
notbias['spanish_text'] = notbias.apply(remove_rt, axis=1)
notbias['spanish_text'] = notbias.apply(remove_mt, axis=1)
notbias['spanish_text'] = notbias.apply(remove_mc, axis=1)
notbias['spanish_text'] = notbias.apply(remove_number, axis=1) 
notbias['spanish_text'] = notbias.apply(remove_urls, axis=1)
notbias['spanish_text'] = notbias.apply(signos, axis=1)


nontoxic['spanish_text'] = nontoxic.apply(demojir, axis=1)
nontoxic['spanish_text'] = nontoxic.apply(remove_user, axis=1)
nontoxic['spanish_text'] = nontoxic.apply(remove_hastag, axis=1)
nontoxic['spanish_text'] = nontoxic.apply(remove_rt, axis=1)
nontoxic['spanish_text'] = nontoxic.apply(remove_mt, axis=1)
nontoxic['spanish_text'] = nontoxic.apply(remove_mc, axis=1)
nontoxic['spanish_text'] = nontoxic.apply(remove_number, axis=1) 
nontoxic['spanish_text'] = nontoxic.apply(remove_urls, axis=1)
nontoxic['spanish_text'] = nontoxic.apply(signos, axis=1)


toxic['spanish_text'] = toxic.apply(demojir, axis=1)
toxic['spanish_text'] = toxic.apply(remove_user, axis=1)
toxic['spanish_text'] = toxic.apply(remove_hastag, axis=1)
toxic['spanish_text'] = toxic.apply(remove_rt, axis=1)
toxic['spanish_text'] = toxic.apply(remove_mt, axis=1)
toxic['spanish_text'] = toxic.apply(remove_mc, axis=1)
toxic['spanish_text'] = toxic.apply(remove_number, axis=1) 
toxic['spanish_text'] = toxic.apply(remove_urls, axis=1)
toxic['spanish_text'] = toxic.apply(signos, axis=1)

```
