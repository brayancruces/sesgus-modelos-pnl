```python
import torch
from transformers import  BertTokenizer
import numpy as np
```

```python
def load_model(model_path):
  return torch.load(model_path)
def setup_tokenizer():
  return BertTokenizer.from_pretrained("pytorch/",do_lower_case=True)
```

```python
def setup(model):
  run_on = 'cuda'
  device = torch.device(run_on)
  model.to(device)
```

```python
def processing(text):
  tokenizer.tokenize(text)
  encoded_doc = tokenizer.encode_plus(text,add_special_tokens=
                True, max_length=115,truncation=True,pad_to_max_length=True)
  input_ids_1=[]
  attention_mask_1=[]
  input_ids_1.append(encoded_doc['input_ids'])
  attention_mask_1.append(encoded_doc['attention_mask'])
  return input_ids_1,attention_mask_1


def make_prediction(model,input, attention_mask):
  with torch.no_grad():
    output = model(torch.tensor(input).to(device), token_type_ids = None, attention_mask = torch.tensor(attention_mask).to(device))
  prediction = 'Negativo' if np.argmax(output.logits.cpu().numpy()).flatten().item() == 1 else 'Positivo'
  print('Input Sentence: ', text)
  print('Predicted Class: ', prediction)
```

```python
run_on = 'cuda'
device = torch.device(run_on)
```

```python
text = 'me alegra que los derechos de los hombres y las mujeres sean tomadas en cuenta'
path='/content/drive/MyDrive/data bert/toxic_model'
model=load_model(path)
tokenizer=setup_tokenizer()
setup(model)
input_ids,attention_mask=processing(text)
make_prediction(model,input_ids, attention_mask)

```
