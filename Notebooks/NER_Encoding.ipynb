```python
dt=df_vf[['spanish_text','bias_name','real_target','index']]
```

```python
dt['len']=dt["spanish_text"].str.split().str.len()
```

```python
def get_array(max,cont,class_N):
  arr_o=[]
  for i in range(0,max):
    if i==cont:
      arr_o.append(class_N)
    else:
      arr_o.append('O')
  return arr_o

def get_all_arr(df):
  arr_gen=[]
  for i in range(0,df.shape[0]):
    print('-',i)
    class_n=df['bias_name'][i]
    print(class_n)
    cont=df['index'][i]
    print(cont)
    max_n=df['len'][i]
    print(max_n)
    classt=' '.join(get_array(max_n,cont,class_n))
    arr_gen.append(classt)
  return arr_gen




 
dt.index=range(0,dt.shape[0])
```

```python
gg=get_all_arr(dt)
```

```python
final=pd.DataFrame(gg, columns=['ner_encoding'])
```

```python
dt['labels']=final['ner_encoding']
```
